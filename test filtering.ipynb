```python
from predict import *
import hw2_evaluation_task2

val_input_image = './hw2_train_val/val1500'
output_label_path = './Test_hbb/stage_{:d}_thres_{:.2f}'
model_path = './models/encurage_detect/stage_{:d}_model.pth'

annotation_path = ''

stages = [0, 10, 20, 30, 40]
conf_thres = [0.0, 0.05, 0.1, 0.15, 0.20, 0.25, 0.30, 0.35]

#stages = [0, 10]
#conf_thres = [0.0, 0.05]

result = []
for stage in stages:
    print("=================== Stage: {:d} ===================".format(stage))
    for cls_conf in conf_thres:
        
        print("Stage: {:d}, Confidence threshold: {:.2f}.".format(stage, cls_conf))
        output_folder = output_label_path.format(stage, cls_conf)
        m_path = model_path.format(stage)
        
        # execute prediction and write files into 'output_folder'
        execution(val_input_image, output_folder, m_path, conf_thres = cls_conf)
        
        # evaluatation  detpath, annotation path
        map1, classaps = hw2_evaluation_task2.main(output_folder, os.path.join(val_input_image, 'labelTxt_hbb'))
        
        print("map: {:.2f}".format( map1))
        result.append((map1, classaps))
```
